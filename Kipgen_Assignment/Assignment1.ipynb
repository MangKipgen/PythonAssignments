```python
#import pandas
import pandas as pd
```

```python
#import csv file and save as dataframe (df)
path = 'medianrentburden.csv'
df = pd.read_csv(path)
```

```python
# all boroughs 
df
```

```python
list(df)
```

```python
# keep only data from columns from 2008-2018
df.drop(['short_name','long_name','2000','2005','2006','2007'],axis=1, inplace=True)
df.head(15)
```

```python
import matplotlib.pyplot as plt
user_input_year = input('Enter year, data available from 20082018 ')
# isolate latest year numbers (2018)
chart = df.plot.scatter(x='Sub-Borough Area',
                y=user_input_year,s=15,c='blue')
#title of chart
plt.suptitle('Median rent burden NYC | ' + user_input_year, fontsize=24, fontweight='bold')
#rotate for readibility
plt.xticks(rotation=90)
#increase chart size
plt.gcf().set_size_inches((15, 7))    

plt.show()
```
